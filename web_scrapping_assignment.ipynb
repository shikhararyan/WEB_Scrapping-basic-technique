```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
import csv

```

```python
url = 'https://sso.agc.gov.sg/Act/TSFA2002?ProvIds=Sc1-#Sc1-'
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}

response = requests.get(url=url, headers=headers)


print(response)
```

```python

soup = BeautifulSoup(response.content, "html.parser")

table = soup.find_all("table", {"class": "p1_1"})
input_str = []


# Print the text
#print(table[1])
tab = table[1].find_all("td",{'class':"sProvP1"})
for text in tab:
    input_str.append(text.get_text())

print(input_str[0:3])

```

```python
with open('output.csv', mode='w', newline='') as file:
    # Create a CSV writer object
    writer = csv.writer(file)
    
    # Write the header row
    writer.writerow(['Name', 'Citizenship', 'Date of Birth'])
    
    # Iterate over the data and write each row
    for item in input_str:
        # Split the string into its components
        name, citizenship, dob = item.split('(')[0].strip(), item.split('(')[1].split(')')[0], item.split(':')[1].strip()[:-1]
        
        # Write the row to the CSV file
        writer.writerow([name, citizenship, dob])
```

```python
name =[]
citizenships =[]
dob= []

for item in input_str:
    # Split the string into its components
    n, c, d = item.split('(')[0].strip(), item.split('(')[1].split(')')[0], item.split(':')[1].strip()[:-1]
    name.append(n)
    citizenships.append(c)
    dob.append(d)
print(name)
print(citizenships)
print(dob)
```
